```python
# ESERCIZIO CORRETTO PER CONTARE LETTERE CON DIZIONARIO

stringa=input('inserisci una parola:')

diz={}

for lettera in stringa:
  if lettera not in diz:
    diz[lettera]=1
  else:
    diz[lettera]+=1

print(diz)
```

```python
# Dato un elenco di parole, contare quante volte appare ciascuna parola utilizzando un ciclo for e un dizionario.
# parole = ["casa", "albero", "cane", "casa", "cane", "gatto", "cane"]


parole = ["casa", "albero", "cane", "casa", "cane", "gatto", "cane"]
dizionario = {}

for parola in parole:
  if parola not in dizionario:
    dizionario[parola]=1
  else:
    dizionario[parola]+=1

for chiave, valori in dizionario.items():
  print(chiave, valori)
```
